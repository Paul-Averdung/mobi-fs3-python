**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical $3 \times 3$ Prewitt filters:

```python
def prewitt_h(img):
    result = np.zeros(img.shape)
    for p in np.ndindex(img.shape):
        neighborhood_sum = 0 
        count = 0
        for q in np.ndindex(3, 3):
            x = p[0] + q[0] - 1
            y = p[1] + q[1] - 1
            if 0 <= x < img.shape[0] and 0 <= y < img.shape[1]:
                if q[1] == 0:
                    neighborhood_sum -= img[x,y]
                elif q[1] == 2:
                    neighborhood_sum += img[x,y]
                count += 1
        if count == 9:
            result[p] = neighborhood_sum / 6
        else:
            result[p] = 0 
    return result

def prewitt_v(img):
    result = np.zeros(img.shape)
    for p in np.ndindex(img.shape):
        neighborhood_sum = 0 
        count = 0
        for q in np.ndindex(3, 3):
            x = p[0] + q[0] - 1
            y = p[1] + q[1] - 1
            if 0 <= x < img.shape[0] and 0 <= y < img.shape[1]:
                if q[0] == 0:
                    neighborhood_sum -= img[x,y]
                elif q[0] == 2:
                    neighborhood_sum += img[x,y]
                count += 1
        if count == 9:
            result[p] = neighborhood_sum / 6
        else:
            result[p] = 0 
    return result
```

## Edge detection

**<a id="#">Task 4.2.1.</a>** Use your results from **Task 4.1.2 â€“ 4.1.3** to compute the *gradient magnitude* of the image:

```python

def magnitude(img):
    result = np.zeros_like(img)
    v_gradient = prewitt_v(img)
    h_gradient = prewitt_h(img)
    for p in np.ndindex(img.shape):
        result[p] = np.sqrt(h_gradient[p]**2 + v_gradient[p]**2)
    return result
        
```
