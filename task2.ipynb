## Intensity clipping

First we define some fixed thresholds $T_1$, $T_2$:

```python
T1 = 0.31
T2 = 0.59
```

### Using `ndarray.clip` method

**<a id="#">Task 2.3.1.</a>** Enter your solution below:

```python
img.clip(T1, T2)
plt.figure(figsize = (8,4))
plt.imshow(img, 'gray')
plt.colorbar()
```

### Not using the `ndarray.clip` method

**<a id="#">Task 2.3.2 (a).</a>** Enter your solution below:

```python
for y in range(img.shape[1]):
    for x in range(img.shape[0]):
        if img[x, y] < T1:
            img[x, y] = T1
        elif img[x, y] > T2:
            img[x, y] = T2
        else:
            img[x, y] = img[x, y]
print(img)
plt.figure(figsize = (8,4))
plt.imshow(img, 'gray')
plt.colorbar()
```

**<a id="#">Task 2.3.2 (b).</a>** Enter your solution below:

```python
for i in numpy.ndindex(img.shape):
    x, y = i[0], i[1]
    if img[x,y] < T1:
        img[x,y] = T1
    elif img[x,y] > T2:
        img[x,y] = T2
    else:
        img[x,y] = img[x,y]
plt.figure(figsize = (8,4))
plt.imshow(img, 'gray')
plt.colorbar() 
```
